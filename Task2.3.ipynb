```python
def switch_it_up(number: int)-> str:
   
    names_of_letters = {
        0: 'Zero',
        1: 'One',
        2: 'Two',
        3: 'Three',
        4: 'Four',
        5: 'Five',
        6: 'Six',
        7: 'Seven',
        8: 'Eight',
        9: 'Nine'
    }
    return (names_of_letters[number])

print(switch_it_up(int(input('Введите цифру'))))
```
